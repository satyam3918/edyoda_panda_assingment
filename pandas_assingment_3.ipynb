```
import numpy as np
import pandas as pd
```

```
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
```

```
df = pd.read_csv(url , sep ='\t')
```

```
df.head()
```

```
df.sort_values(by ='item_name' , ascending = True).reset_index(drop= True)
```

```
count_veggi_salad = np.count_nonzero(df.item_name == 'Veggie Salad Bowl')
print(count_veggi_salad)
```

```
canned_soda = np.count_nonzero((df.item_name == 'Canned Soda') & (df.quantity > 1))
```

```
print(canned_soda)
```
